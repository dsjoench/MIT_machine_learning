```python
import numpy as np
import math

p1 = p2=  0.5
mu = -3
mu2=2
sigma =sigma2= 4

x = [0.2,-0.9,-1,1.2,1.8]
p = []

mu_num = []
print("E step: ")

for i in x:
    com =  p1*(1/(2*math.pi*sigma))*np.exp(-1/(2*sigma)* abs(i - mu)**2)
    
    com_2 = p2*(1/(2*math.pi*sigma2))*np.exp(-1/(2*sigma2)* abs(i - mu2)**2)
    ans = (com/(com+com_2))
    p.append(ans)
    print(ans)
    mu_num.append(ans*i)
print("")
print("M Step:")
print(f'mu: {np.sum(mu_num)/ np.sum(p)}')
print(f'new_p : {np.mean(p)} ')
print(f'new simga: {np.sum(p * (abs(x - np.sum(mu_num)/ np.sum(p))**2)) /(1*np.sum(p))             }')
    
    



```
